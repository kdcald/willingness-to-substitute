```python
sub_dairy_dict = {'sub_dairy_cottage_cheese' : 'Cottage cheese',
                  'sub_dairy_cheese_other' : 'Cheese, other',
                  'sub_dairy_cheese_cheddar' : 'Cheese, cheddar',
                  'sub_dairy_cheese_american' : 'Cheese, American',
                  'sub_dairy_sour_cream' : 'Sour cream',
                  'sub_dairy_cream' : 'Cream',
                  'sub_dairy_dressing' : 'Dressing',                  
                  'sub_dairy_butter' : 'Butter',        
                  'sub_dairy_ice_cream' : 'Ice cream',
                  'sub_dairy_yogurt' : 'Yogurt',
                  'sub_dairy_milk' : 'Milk'}

sub_order = ['Already substitute', 'Very likely', 'Likely', 
              'Neutral', 'Unlikely', 'Very unlikely', 'Definitely would not']
```

```python
data = []

for x in list(sub_dairy_dict.keys()):
    data.append((df[x].value_counts(normalize=True).reindex(sub_order)*100).fillna(0).values.tolist())
```

```python
top_labels = ['Already<br>substitute', 'Very<br>likely', 'Likely', 
              'Neutral', 'Unlikely', 'Very<br>unlikely', 'Definitely<br>would not']

colors = ['rgb(38, 70, 83)', 'rgb(42, 157, 143)', 'rgb(42, 157, 143)',
          'rgb(233, 196, 106)', 'rgb(244, 162, 97)',
          'rgb(244, 162, 97)', 'rgb(231, 111, 81)']

x_data_labels = np.array(list(np.around(np.array(data))), dtype=int)

x_data = data

y_data = list(sub_dairy_dict.values())

fig = go.Figure()

for i in range(0, len(x_data[0])):
    for xd, yd in zip(x_data, y_data):
        fig.add_trace(go.Bar(
            x=[xd[i]], y=[yd],
            orientation='h',
            marker=dict(
                color=colors[i],
                line=dict(color='rgb(248, 248, 249)', width=1)
            )
        ))

fig.update_layout(
    title='How likely are participants to substitute dairy products?',
    xaxis=dict(
        showgrid=False,
        showline=False,
        showticklabels=False,
        zeroline=False,
        domain=[0.15, 1]
    ),
    yaxis=dict(
        showgrid=False,
        showline=False,
        showticklabels=False,
        zeroline=False,
    ),
    barmode='stack',
    paper_bgcolor='rgb(248, 248, 255)',
    plot_bgcolor='rgb(248, 248, 255)',
#    margin=dict(l=120, r=10, t=140, b=80),
    showlegend=False,
)

annotations = []

for yd, xd in zip(y_data, x_data_labels):
    # labeling the y-axis
    annotations.append(dict(xref='paper', yref='y',
                            x=0.14, y=yd,
                            xanchor='right',
                            text=str(yd),
                            font=dict(family='Arial', size=12,
                                      color='rgb(67, 67, 67)'),
                            showarrow=False, align='right'))
    # labeling the first percentage of each bar (x_axis)
    annotations.append(dict(xref='x', yref='y',
                            x=xd[0] / 2, y=yd,
                            text=str(xd[0]) + '%',
                            font=dict(family='Arial', size=10,
                                      color='rgb(248, 248, 255)'),
                            showarrow=False))
    # labeling the first Likert scale (on the top)
    if yd == y_data[-1]:
        annotations.append(dict(xref='x', yref='paper',
                                x=xd[0] / 2, y=1.1,
                                text=top_labels[0],
                                font=dict(family='Arial', size=10,
                                          color='rgb(67, 67, 67)'),
                                showarrow=False))
    space = xd[0]
    for i in range(1, len(xd)):
            # labeling the rest of percentages for each bar (x_axis)
            annotations.append(dict(xref='x', yref='y',
                                    x=space + (xd[i]/2), y=yd,
                                    text=str(xd[i]) + '%',
                                    font=dict(family='Arial', size=10,
                                              color='rgb(248, 248, 255)'),
                                    showarrow=False))
            # labeling the Likert scale
            if yd == y_data[-1]:
                annotations.append(dict(xref='x', yref='paper',
                                        x=space + (xd[i]/2), y=1.1,
                                        text=top_labels[i],
                                        font=dict(family='Arial', size=10,
                                                  color='rgb(67, 67, 67)'),
                                        showarrow=False))
            space += xd[i]

fig.update_layout(annotations=annotations)

fig.show()
```

```python
top_labels = ['Extremely<br>familiar', 'Moderately<br>familiar', 'Somewhat<br>familiar', 'Slightly<br>familiar', 'Not at<br>all familiar']

colors = ['rgb(38, 70, 83)', 'rgb(42, 157, 143)', 'rgb(148, 210, 189)',
          'rgb(233, 196, 106)', 'rgb(244, 162, 97)',
          'rgb(244, 162, 97)']

x_data_labels = np.array(list(np.around(np.array(data))), dtype=int)

x_data = data

y_data = list(familiarity_dict.values())

fig = go.Figure()

for i in range(0, len(x_data[0])):
    for xd, yd in zip(x_data, y_data):
        fig.add_trace(go.Bar(
            x=[xd[i]], y=[yd],
            orientation='h',
            marker=dict(
                color=colors[i],
                line=dict(color='rgb(248, 248, 249)', width=1)
            )
        ))

fig.update_layout(
    title='How familiar are you with plant-based alternatives?',
    xaxis=dict(
        showgrid=False,
        showline=False,
        showticklabels=False,
        zeroline=False,
        domain=[0.15, 1]
    ),
    yaxis=dict(
        showgrid=False,
        showline=False,
        showticklabels=False,
        zeroline=False,
    ),
    barmode='stack',
    paper_bgcolor='rgb(248, 248, 255)',
    plot_bgcolor='rgb(248, 248, 255)',
#    margin=dict(l=120, r=10, t=140, b=80),
    showlegend=False,
)

annotations = []

for yd, xd in zip(y_data, x_data_labels):
    # labeling the y-axis
    annotations.append(dict(xref='paper', yref='y',
                            x=0.14, y=yd,
                            xanchor='right',
                            text=str(yd),
                            font=dict(family='Arial', size=12,
                                      color='rgb(67, 67, 67)'),
                            showarrow=False, align='right'))
    # labeling the first percentage of each bar (x_axis)
    annotations.append(dict(xref='x', yref='y',
                            x=xd[0] / 2, y=yd,
                            text=str(xd[0]) + '%',
                            font=dict(family='Arial', size=10,
                                      color='rgb(248, 248, 255)'),
                            showarrow=False))
    # labeling the first Likert scale (on the top)
    if yd == y_data[-1]:
        annotations.append(dict(xref='x', yref='paper',
                                x=xd[0] / 2, y=1.1,
                                text=top_labels[0],
                                font=dict(family='Arial', size=10,
                                          color='rgb(67, 67, 67)'),
                                showarrow=False))
    space = xd[0]
    for i in range(1, len(xd)):
            # labeling the rest of percentages for each bar (x_axis)
            annotations.append(dict(xref='x', yref='y',
                                    x=space + (xd[i]/2), y=yd,
                                    text=str(xd[i]) + '%',
                                    font=dict(family='Arial', size=10,
                                              color='rgb(248, 248, 255)'),
                                    showarrow=False))
            # labeling the Likert scale
            if yd == y_data[-1]:
                annotations.append(dict(xref='x', yref='paper',
                                        x=space + (xd[i]/2), y=1.1,
                                        text=top_labels[i],
                                        font=dict(family='Arial', size=10,
                                                  color='rgb(67, 67, 67)'),
                                        showarrow=False))
            space += xd[i]

fig.update_layout(annotations=annotations)

fig.show()
```
